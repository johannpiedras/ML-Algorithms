## Guassian Naive Bayes Algorithmn



Bayes Theorem provides a way that we can calculate the probability of a piece of data belonging to a given class, given our prior knowledge. 

P( class | data ) = P( data | class ) * (P(class)) / P(data) <br>
Where P( class | data ) is the probability of class given the provided data. <br>

Another way to think about this in terms of this algorithmn is: 

Posterior Probability = (Likihood * Prior) / Evidence 



**Steps in our algorithmn** <br>
Step 1: Separate By Class. \
Step 2: Summarize Dataset. \
Step 3: Summarize Data By Class. \
Step 4: Gaussian Probability Density Function. \
Step 5: Class Probabilities 

```python
class GaussNB():
    
    
    def fit(self, X, y, epsilon = 1e-3):
        # We can create a dictionary object where each key is the class value and 
        # then add a list of all the records as the value in the dictionary.
        
        self.likihoods = dict() # want to keep track what likihood is from what observation
        self.priors = dict() #
        
        self.K = set(y.astype(int)) # this is for defining classes, creating an ordered list of iterables without duplicates
        
        for k in self.K:
            
            X_k  = X[y==k,:] 
            # The first time, k == 0, so it gets the weight and height of women
            # The second time, k == 1, it gets the weight and height of men
            
            self.likihoods[k] = {"mean" : X_k.mean(axis=0), "cov": X_k.var(axis = 0) + epsilon } 
            # storing as a dictionary
            # so there is a dictionary which is likihood, and its nested with another dictionary of mean and cov
            
            self.priors[k] = len(X_k)/len(X)  #  class / n
        
    
    # Now we get to the prediction
    def predict(self, X):
        
        N,D = X.shape # a way to create two variables quickly. Since shape has 2 output N will be row, D columns
        P_hat = np.zeros((N,len(self.K))) # creating shape of P_hat, just fulling with zeros at the moment
        
        for k, l in self.likihoods.items(): #looking through two different axis, k to the rows, l to the columns
            # doing calculation for every single obsercation
            # we calculate the PDF
            # then we add the log of the priors
            P_hat[:,k] = mvn.logpdf(X,l["mean"], l["cov"]) + np.log(self.priors[k]) # getting a function that gives us the probablity density
            
        return P_hat.argmax(axis=1) # maximum postier estimator
        
        
```

```python
# defining an instance of our class... a fancy name for variable
gnb = GaussNB()
gnb_g = GaussNB()
```

```python
gnb.fit(X,y)
y_hat = gnb.predict(X)
```

```python
gnb_g.fit(X_g,y_g)
y_hat_g = gnb.predict(X_g)
```

```python
# function definition 
def accuracy(y,y_hat):
    return np.mean(y == y_hat)
```

```python
print(f"Accuracy : {accuracy(y,y_hat):0.3f}")
```

```python
print(f"Accuracy : {accuracy(y_g,y_hat_g):0.3f}")
```

### Gauss Bayes 

```python
class GaussBayes(): # not naive 
    
    
    def fit(self, X, y, epsilon = 1e-3):

        
        self.likihoods = dict() # 
        self.priors = dict() #     
        self.K = set(y.astype(int)) # 
        
        for k in self.K:   
            X_k = X[y==k,:]
            N_k, D = X_k.shape
            mu_k = X_k.mean(axis = 0)
            self.likihoods[k] = {"mean" : X_k.mean(axis=0), "cov": (1/(N_k - 1))*np.matmul((X_k - mu_k).T, X_k - mu_k) + epsilon*np.identity(D) } 
            # when multiplying a matrix by its transpose is what? What does that do
            self.priors[k] = len(X_k)/len(X)  
        
    
    # Now we get to the prediction
    def predict(self, X):
        N,D = X.shape # 
        P_hat = np.zeros((N,len(self.K))) # this is a tuple
        
        for k, l in self.likihoods.items(): 
 
            P_hat[:,k] = mvn.logpdf(X,l["mean"], l["cov"]) + np.log(self.priors[k]) # getting a function that gives us the probablity density
            
        return P_hat.argmax(axis=1) # maximum postier estimator
        
        
```

```python
gb = GaussBayes()
```

```python
gb.fit(X,y)
```

```python
y_hat = gb.predict(X)
```

```python
print(f"Accuracy : {accuracy(y,y_hat):0.3f}")
```
