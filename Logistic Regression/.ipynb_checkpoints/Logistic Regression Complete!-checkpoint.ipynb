$\phi_j (x_i) = e^{-\beta \vert \vert x_i -x_j \vert \vert ^2}$

where $\beta = \frac{1}{2 \sigma^2}$

# Logistic Regression:



the "Hypothesis":

$h(x_i)=\sum_{j=1}^D w_j x_{ij} +b$   This is a dot product 

$=w^t x_1 +b =w^T X_i$

if $w^T x +b =0$  x is on the desicion boundary


To get predictions:

Targets:  $y_i \in {0,1}$ 

The targets obey a Bernoulli distribution such that $y_i$ ~ $B(p_i)$  and $p(y_i=1\vert x_i)$

We need to calculate the "odds" to get p  such that $odds(p)=\frac{p}{1-p}$  

The log of odds $logits(p)=ln(\frac{p}{1-p})$

the inverse function is derived:$ h=ln(\frac{p}{1-p})$ thus $e^h=\frac{p}{1-p}$

Thus $p=\frac{e^h}{1+e^h}$

and thus our "Sigmoid" Function $p=\frac{1}{1+e^{-h}}$ and $$\sigma(h)=\frac{1}{1+e^h}$$





Assumptions about Data:

Idential Gaussian distribution to both clusters such that:

$p(x\vert y=0)=\frac{1}{\sqrt (2\pi )^D \vert \Sigma \vert} e^{frac{1}{2}(x-\mu_0)^T \Sigma^-1 (x-\mu_o)}$

$p(x\vert y=1)=\frac{1}{\sqrt (2\pi )^D \vert \Sigma \vert} e^{frac{1}{2}(x-\mu_1)^T \Sigma^-1 (x-\mu_1)}$

Then the probability that y=1 give x,  is given by 

$p(y=1\vert x) =\frac{p(x\vert y=1)p(y=1)}{p(x}$

$p(y=1\vert x) =\frac{p(x\vert y=1)p(y=1)}{p(x\vert y=1)p(y=1)+p(x\vert y=0)p(y=0)}* \frac{\frac{1}{p(x\vert y=1)p(y=1)}}{\frac{1}{p(x\vert y=1)p(y=1)}}$

$p(y=1\vert x)=\frac{1}{1+\frac{p(x\vert y=0)p(y=0)}{p(x\vert y=1)p(y=1)}}$

take the natural log of the thing in the denominator:

$=ln(p(x\vert y=0)p(y=0))-ln(p(x\vert y=0)p(y=0))+\alpha$   where $\alpha= ln(\frac{1-p}{p})$

$=-\frac{1}{2} ln[(2\pi)^D\vert\Sigma \vert] -\frac{1}{2}(x-\mu_o)^T\Sigma^{-1}(x-\mu_o) +\frac{1}{2}ln[(2\pi)^D\vert\Sigma \vert]+\frac{1}{2}(x-\mu_1)^T\Sigma^{-1}(x-\mu_1)+\alpha$

$=\frac{1}{2}{-x^T\Sigma^{-1}x+x^T\Sigma^{-1}+\mu_o^T\Sigma^{-1}x-\mu_o^T \Sigma^{-1}\mu_0}$

but we know that $x^T\Sigma^{-1}=(\Sigma^{-1})^Tx=\mu^T\Sigma^{-1}\mu_o+\alpha$

So, 

$=\mu_o^T \Sigma ^{-1}x+\mu_1^T\Sigma^{-1}x+\frac{1}{2}(\mu_1^T \Sigma^{-1}\mu_1-\mu_o^T\Sigma^{-1} \mu_o)+\alpha$

$=(\mu_o-\mu_1)^T\Sigma^{-1} x+\frac{1}{2}(\mu_1^T \Sigma^{-1}\mu_1-\mu_o^T\Sigma^{-1}\mu_o)+\alpha$

Define:  $w=\Sigma^{-1}(\mu_1-\mu_0)$

$b=\frac{1}{2}(\mu_o^T\Sigma^{-1}\mu_o-\mu_1^T \Sigma^{-1}\mu_1)-\alpha$

returning to the log of $that$ :

$=-(\mu_1-\mu_o)^T\Sigma^{-1}x-[frac{1}{2}(\mu_o^T\Sigma^{-1}\mu_o-\mu_1^T \Sigma^{-1}\mu_1)-\alpha]$

$-w^Tx-b = -(w^tx+b)$  and thus, $p(y=1\vert x)=\frac{1}{1+e^{-w^Tx_b)}}$ and $w^Tx+b=h = \mathbf{logit}(p)$

And the probability $$\hat p_i = \frac{1}{1+e^{-(w^T x_i +b)}}$$

the prediction is made by setting a threshold in an "Indicator Function":

$$\hat y_i = \mathbf {ind} (\hat p_i \geq t)$$

The Objective Function:  

$$\mathcal J =- \frac{1}{N} \sum_{i=1}^N y_i ln \hat p_i +(1-y_i)ln (1-\hat p_i)$$

And it's called the "Cross Entropy Objective"  a convex function.




This is the function we have to differentiate to do Gradient Descent on this.

$\mathcal J(\hat p) =- \frac{1}{N} \sum_{i=1}^N y_i ln \hat p_i +(1-y_i)ln (1-\hat p_i)$

$p(h)=\frac{1}{1+e^{-h}}$

$\frac{\partial \mathcal J}{\partial \hat w}=\frac{\partial \mathcal J}{\partial \hat p} \frac{\hat p}{d h} \frac{\partial h}{\partial w}$

$h(w,b)= w^T x_i +b$   

and, 

$\frac{\partial \mathcal J}{\partial \hat p_i} = -\frac {1}{N} \sum_{i=1}^N \frac{y_i}{p_i}-\frac{1-y_i}{1-p_i}$ 

$\frac{\hat p_i}{d h_i}=\hat p(1-\hat p)$

$\frac{\partial h_i}{\partial w_j}= x_{ij}$

$\frac{\partial \mathcal J}{\partial \hat w}=-\frac {1}{N} \sum_{i=1}^N (\frac{y_i}{\hat p_i}-\frac{1-y_i}{1- \hat p_i})\hat p(1-\hat p) x_{ij}$

$\frac{\partial \mathcal J}{\partial \hat w}=-\frac {1}{N} \sum_{i=1}^N x_{ij} (y_i -\hat p_i) $

$\frac{\partial \mathcal J}{\partial \hat w}=\frac {1}{N} \sum_{i=1}^N x_{ij} ( \hat p_i -y_i) $

Therefore,  the Gradient is defined as:

$$\nabla \mathcal J = \frac{1}{n} X^T(\hat p - y)$$ where p, y are 1 by N column vectors and X is a N by D matrix and $\frac{\partial \mathcal J }{\partial b}=\frac{1}{N} \sum_{i=1}^{N} \hat p_i - y_i$

Implementation in Python:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
%matplotlib inline
```

```python
cmap_bold = ListedColormap(["#FF0000","#0000FF"])
cmap_light = ListedColormap(["#FFBBBB","#BBBBFF"])
```

```python
def sigmoid (h):
    return  1/(1 + np.exp(-h))

def cross_entropy(y,p_hat):
    return -(1/len(y))*np.sum(y*np.log(p_hat)+(1-y)*np.log(1-p_hat))

def accuracy(y, y_hat):
    return np.mean(y == y_hat)
```

```python
class LogisticRegression():
    
    def __init__(self, thresh = 0.5):
        self.thresh = thresh
        self.w = None
        self.b = None
        
    
    def fit(self, X, y, eta = 1e-3, epochs = 1e3, show_curve = False):
        epochs = int(epochs)
        N, D = X.shape
        
        self.w = np.random.randn(D)
        self.b = np.random.randn(1)
        
        J = np.zeros(epochs)
        
        for epoch in range(epochs):
            p_hat = self.__forward(X)
            J[epoch] =cross_entropy(y,p_hat)
            self.w -= eta*(1/N)*X.T@(p_hat - y)
            self.b -= eta*(1/N)*np.sum(p_hat - y)
            
        if show_curve:
            plt.figure()
            plt.plot(J)
            plt.xlabel("epochs")
            plt.ylabel("$\mathcal{J}$")
            plt.title("Training Curve")
            plt.show()
            
    def __forward(self,X):
        return sigmoid(X@self.w+self.b)
            
    def predict(self, X):
        return (self.__forward(X) >= self.thresh).astype(np.int32)
        
```

```python
def main():
    N = 1000
    D = 2
    
    X0 = np.random.randn((N//2),D) + np.array([1, 1])
    X1 = np.random.randn((N//2),D) + np.array([-1, -1])
    X = np.vstack((X0, X1))
    
    y = np.array([0]*(N//2) + [1]*(N//2))
    
    
    plt.figure()
    plt.scatter(X[:,0], X[:,1], c = y, alpha = 0.5)
    plt.show()
    
    
    df=pd.DataFrame(X)
    df=pd.concat([df, pd.DataFrame(y)], axis=1)
    df.to_csv("LogReg1.csv")
    
    log_reg = LogisticRegression()
    log_reg.fit(X, y, eta = 1e-1, show_curve = True)
    y_hat = log_reg.predict(X)
    
    print(f"Training Accuracy: {accuracy(y, y_hat):0.4f}")
    
    x1 = np.linspace(X[:,0].min() - 1, X[:,0].max() + 1, 1000)
    x2 = -(log_reg.b/log_reg.w[1]) - (log_reg.w[0]/log_reg.w[1])*x1
    
    
    plt.figure()
    plt.scatter(X[:,0], X[:,1], c = y, alpha = 0.5)
    plt.plot(x1, x2, color = "#000000", linewidth = 2)
    plt.show()
    
    xx1, xx2 = np.meshgrid(x1, x1)
    Z = log_reg.predict(np.c_[xx1.ravel(),xx2.ravel()]).reshape(*xx1.shape)
    
    plt.figure()
    plt.pcolormesh(xx1, xx2, Z, cmap = cmap_light)
    plt.scatter(X[:,0], X[:,1], c = y, cmap = cmap_bold)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    plt.show()
```

```python
if __name__=="__main__":
    main()
```

# What does it All Mean?

$y_i$~$B(p_i)$   Our obsevations are independent and follow the Bernoulli distribution 

then our Likelihood function:

$\mathcal L (y| X, w) = \prod_{i=1}^N p_i^{y_i} (1-p_i)^{1-y_i}$

the probability is $ p_x(x)=p^x(1-p)^{1-x}$

So to maximize the likelihood, we maximize the log Likelihood

so $l= log \mathcal L $  is given by 

$ l( y | X,w) = \sum_{i=1}^N \mathbf {ln}( p_i^{y_i} ) +\mathbf {ln}((1-p_x)^{1-y_i})=   \sum_{i=1}^N y_i \mathbf{ln} p_i + (1-y_i) \mathbf {ln} (1-p_i)=-N \mathcal J$ 

so for maxima/minima and optimization purposes:$$ \mathcal J = -l$$

let $\sigma$ be the sigmoid function and if $\hat y_i=w^Tx_i +b$  

then $\hat p_i = \sigma (e^T x_i +b)$

and $\hat p_i = \sigma(h)=\frac{1}{1+e^{-h_i}}$

### Non-linear features:

$h=f(X)+\epsilon$

$\hat h =\hat f (X)$

f can be expressed as a linear combination of basis vectors:{$\phi_i,...,\phi_p$} where $f\in \mathcal F(\mathbb{R}^D \mathbb{R})$

so now, $\hat y_i = w^T \phi (x_i)+b$  where $\phi(x)$ is our "engineered" feature and function  

So to keep from overfitting the boundary we add "L2" regularization:



$\mathcal J =- \frac{1}{N} \sum_{i=1}^N y_i ln \hat p_i +(1-y_i)ln (1-\hat p_i)+ \frac{\lambda}{2N}\sum_{j=1}^D w_j^2$

Let $\mathcal J=-l$ and $\mathcal L = e^l$

then, $\mathcal L = e^{-\mathcal J} = e^{\sum_{i=1}^N y_i ln \hat p_i +(1-y_i)ln (1-\hat p_i)+ \frac{\lambda}{2N}\sum_{j=1}^D w_j^2}$

$= \prod_{i=1}^N e^{ y_i ln \hat p_i +(1-y_i)ln (1-\hat p_i)} \prod_{j=1}^{P} e^{\frac{\lambda}{2} w_j^2}$

$= \prod_{i=1}^N  y_i ln \hat p_i^{y_i}(1-y_i)^ {(1-\hat p_i)} \prod_{j=1}^{P} e^{\frac{\lambda}{2} w_j^2}$

$$\mathcal L \propto p(y|X,w)p(w)\propto p(w|X,y)$$
